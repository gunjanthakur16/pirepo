```python
def call(a,b,d): # Positional Arguments
    print(a,b,d)
    return

age=12
studentname="Rahul"
classname="BECSEAI"

call(b="Amit",d="BE-CSE-AI",a=10)  #Keyword Arguments
```

```python
def call(a=12,b="Rahul",d="BE-CSE"): # Default Arguments
    print(a,b,d)
    return

call(1,2,3)  #Keyword Arguments
```

```python
def call(*arg):  # Variable number of arguments
    for i in arg:
        print(i)
    return

#call(1)
call(1,2)
call(1,2,3)
call(1234,234,456,456,546,456)
call(3424,234,456,546,54,4324,234,324,234,23,423,4234)
```

```python
# Fibonacci Series
def fib(n):
    a,b=0,1
    while a<=n:
        print(a)
        a,b=b,a+b
        
```

```python
fib(30)
```

```python
def fib(n):
    a,b=0,1
    while a<=n:
        yield a
        a,b=b,a+b
```

```python
mygen=fib(30)
```

```python
for i in mygen:
    print(i)
```
